## hamming distance

```python
%load_ext autoreload
%autoreload 2
from run_circuit import IsingChainSimulation,TiltedIsingChain
import matplotlib.pyplot as plt
L = 7
J = 2
hy = 0.37
hz = 1
epsilon = 0.01
initial_state = '1010101'
chain = TiltedIsingChain(L,hz,J,hy,epsilon)
# chain.correlation_one_site(initial_state, 2,40)
distance = chain.hamming_distance(initial_state, 40)
plt.plot(distance)
plt.xlabel('Time (cycles)') 
plt.ylabel('Hamming Distance')
plt.title('Hamming Distance over Time')
plt.grid()
plt.show()
```
